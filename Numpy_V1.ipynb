### 1. Why Numpy is Faster than Python Lists?
NumPy is a Python fundamental package which used for efficient manipulations and operations on High-level mathematical functions, Multi-dimensional arrays, Linear algebra, Fourier Transformations, Random Number Capabilities, etc. It provides tools for integrating C, C++ and Fortran code in Python. NumPy is mostly used in Python for scientific computing.

**NumPy Arrays are faster than Python Lists because of the following reasons:**
1. An array is a collection of homogeneous data-types which are stored in contagious memory locations, on the other hand, a list in Python is collection of heterogeneous data types stored in non-contagious memory locations.
1. The NumPy package breakdowns a task into multiple fragments, and then processes all the fragments parallelly.
1. The NumPy package integrates C, C++ and Fortran codes in Python, these programming languages have very less execution time as compared to python.

### **1. numpy.arange()**
***arange([start,] stop[, step,][, dtype])*** : <br>*italicized text*
>Returns an array with evenly spaced elements as per the interval. The interval mentioned is half opened i.e. [Start, Stop)<br>

***Parameters :***<br>
>**start :** [optional] start of interval range. By default start = 0<br>
**stop  :** end of interval range<br>
**step  :** [optional] step size of interval. By default step size = 1 <br>
**dtype :** type of output array
image.png

```
#1. numpy array using arange() function
#using default parameters
a = np.arange(10)
print(a)
```

```
#by usung parameters
a = np.arange(start=10,stop=50,step=4)
print('Array a is :')
print(a)

print('\nArray b is :')
b= np.arange(start=10, stop=50, step=4, dtype='float')
print(b)
```

***Parameters :***<br>

### **3. Common Arrays**
> #### **1. numpy.ones()**
>>***numpy.ones(shape, dtype = None, order = â€˜Câ€™) :*** : <br>
Return a new array of given shape and type, with ones.

>>***Parameters :***<br>
>>**shape :** integer or sequence of integers<br>
**order  :** C_contiguous or F_contiguous.
<br>C-contiguous order in memory(last index varies the fastest).<br>C order means that operating row-rise on the array will be slightly quicker.<br>FORTRAN-contiguous order in memory (first index varies the fastest).<br>F order means that column-wise operations will be faster. <br>
**dtype :** [optional, float(byDeafult)] Data type of returned array.  

```
#print ones array
a = np.ones(2, dtype = int) 
print("Matrix a : \n", a) 
```

```
#print 2-dimensional matrix
a = np.ones([7,4], dtype = int) 
print('Matrix a with 7-rows and 4 columns:')
print(a)
```

> #### **2. numpy.zeros()**
>>***numpy.zeros(shape, dtype = None, order = â€˜Câ€™) :*** : <br>
Return a new array of given shape and type, with zeros.

>>***Parameters :***<br>
>>**shape :** integer or sequence of integers<br>
**order  :** C_contiguous or F_contiguous.
<br>C-contiguous order in memory(last index varies the fastest).<br>C order means that operating row-rise on the array will be slightly quicker.<br>FORTRAN-contiguous order in memory (first index varies the fastest).<br>F order means that column-wise operations will be faster. <br>
**dtype :** [optional, float(byDeafult)] Data type of returned array. 

```
#print zeros array
a = np.zeros(2, dtype = int) 
print("Matrix a : \n", a) 
```

```
#print 2-dimensional matrix
a = np.zeros([7,4], dtype = int) 
print('Matrix a with 7-rows and 4 columns:')
print(a)
```

> #### **3. numpy.eye()**
>>***numpy.eye(R, C = None, k = 0, dtype = type <â€˜floatâ€™>) :*** : <br>
Return a matrix having 1â€™s on the diagonal and 0â€™s elsewhere w.r.t. k.

>>***Parameters :***<br>
>>**N :** Number of rows.<br>
**CM  :** [optional] Number of columns; By default N=M <br>
**k   :** [int, optional, 0 by default]
<br>Diagonal we require; k>0 means diagonal above main diagonal or vice versa.<br>
**dtype :** [optional, float(byDeafult)] Data type of returned array. <br><br>
>>***Returns :***<br>array of shape, R x C, an array where all elements 
are equal to zero, except for the k-th diagonal, 
whose values are equal to one.

```
a = np.eye(2, dtype = float) 
print("Matrix a : \n", a) 
```

```
a = np.eye(N=8,M=6, k=2, dtype='int8', order='C')
print('Matrix with diagonal k=2:')
print(a)

a = np.eye(N=8,M=6, k=-2, dtype='int8', order='C')
print('\nMatrix with diagonal k=-2:')
print(a)
```

> #### **4. numpy.identity()**
>>***numpy.identity(n, dtype = None) :*** : <br>
Return a identity matrix i.e. a square matrix with ones on the main daignol.

>>***Parameters :***<br>
>>**n   :** [int] Dimension n x n of output array.<br>
**dtype :** [optional, float(byDeafult)] Data type of returned array. 

```
a = np.identity(4, dtype='int8')
print(a)
```

> #### **5. numpy.diag()**
>>***numpy.diag(v,k) :*** : <br>
Return a identity matrix i.e. a square matrix with ones on the main daignol.

>>***Parameters :***<br>
>>**v   :** [int] Arraty vector can be 2-dimensional.
**k :** kth diagonal. 

```
a = np.diag([1,2,3,4,5], -3)
print(a)

#we can also extract diagonal element from the given array.
print('\nDiagonal of above matrix is :\n',np.diag(a,-3))
```

## **Get Dimension of an Array**

```
#import and initialize an array using NumPy
import numpy as np 
a = np.array([1,2,3,4,5])
b = np.array([9,8,7,6,5])

#get dimensions of an array
print("'.ndim' attributes gives the dimension of an array")
print('\nDimension of array a is : ', end='')
print(a.ndim)

c = np.array([[1,2,3,4],[4,5,6,7]])
print('Dimenssion of array c is : ', end='')
print(c.ndim)
```

## **Specify the Data type While initializing an Array**

```
#specify Data type of an array while initilaization
a = np.array([1,2,3], dtype='int8')
print('Integer Array\n',a, end='\n\n')

a = np.array([1,2,3], dtype='float')
print('Float Array\n',a, end='\n\n')

a = np.array([1,2,3], dtype='complex')
print('Complex Array\n',a, end='\n\n')

a = np.array([1,2,3], dtype='U')
print('Unicode Array\n',a, end='\n\n')
```

## **Extract Data Type of a Given Array**

```
#get the data type of an Array 
print(a)
print(a.dtype, end='\n\n')

b = np.array([1.2,3.4,5.])
print(b)
print(b.dtype, end='\n\n')

c= np.array(['A','B','C','D'])#unicode
print(c)
print(c.dtype, end='\n\n')

d = np.array([1+2j])
print(d)
print(d.dtype)
```

## **Find out the Shape of an Array**

```
#get shape of an array
a = np.array([1,2])
print(a)
print ('Shape of array a : ', a.shape, end='\n\n')

b = np.array([[1,2,3],[4,5,6]])
print(b)
print('Shape of Array b : ', b.shape,end='\n\n')

c = np.array([[[1,2,6],[2,3,6]],[[3,4,6],[4,5,6]]])
print(c)
print('Shape of Array c : ', c.shape)
```

## **Find out the Total Number of Element in given Array**

```
#get total elements present in an array
a = np.array([[[1,2,3,4],[5,6,7,8]],[[0,9,8,7],[6,5,4,3]]])
print(a)

print('\nTotal Number of Elements in array a : ',a.size)
print('Length of each item in bytes is :', a.itemsize)
print('Total bytes consumed by an array is : ',a.nbytes)
```
