# **Arguments in Python:**
In Python you can define function that takes a variable number of arguments, i.e. we can call a function with different types of arguments/parameters. Arguments can be passed to user defined function or the built-in functions based on their definitions, let’s look at the example for argument-based user defined function.


```
def sum (a,b):
  print('Sum is :',a+b)

sum(56,86)
```

In above example whatever values we pass to *sum(a, b)* are the arguments, here we have 56 and 86 as an arguments.

Python has provided us the multiple ways to pass an argument to the function be it user defined or built-in function, we will look into the user defined and built-in functions later for now let’s look at some these argument passing methods, there are two different arguments assigning methods as follows:


1.	Fixed Function Arguments.
2.	Variable Function Arguments.<br>
a.	Python Default arguments.<br>

c.	Python Arbitrary Argument:



## 1. Fixed Function Arguments:
In Fixed function arguments all the parameters specified in the function definition must be used while calling the function else it will raise an exception.

```
def sum(a,b):
  print('Sum is : ',a+b)

sum(5)
```

In above example we defined function *sum(a, b)* with two parameters, so to call this function successfully we need to make a call with two arguments, but as you can see we only used single argument to call the *sum(a, b)* function which was *sum(5)*. Thus, due to the argument mismatch in function definition and function call we now have an error raised as:

>><h3>Error:</h3>
>TypeError: sum() missing required positional argument: ‘b’

>><h3>Note: </h3>
>For fixed function argument function, the function definition and function call must have same number of arguments, that too in required ORDER.


## 2.	Variable Function Arguments:
Python also gives us an option where we can pass the variable number of arguments to the function unlike the fixed function arguments. In variable function arguments we can define the functions with some default data and other methods, let’s look into these variable function argument methods.

> <h3>a.	Python Default arguments:</h3>Function parameters can have default values assigned to them, let’s look at an example for it as follows:

```
def sum(a, b=85):           #function with default parameter b=85
  print('Sum is :', a+b)    #prints sum of a and b

sum(51)                     #Variable function calling, b will be dafault i.e. b=85
sum(12,)
sum(123,523)                #default variable b will be replaced by second parameter : 523
```

In above example we have 3 different types of function calls, and three different ways with which we can call the user defined function.
1.	In first scenario, we passed only one argument so by default function used the default value for b which is b=85.
2.	In second case, we added comma indicating the two different parameters and kept second parameter which is b as blank, so function used default value as b=85.
3.	In third case, both the parameters are manually passed thus replacing the default parameter. So, the default values are used only if there are no parameters passed.

>><h3>Note: </h3>
Any number of arguments in a function can have a default value. But once we have a default argument, all the arguments to its right must also have default values This means to say, non-default arguments cannot follow default arguments For example, if we had defined the function header above as:

***1. all the arguments to its right must also have default values.***<br>
***2. non-default arguments cannot follow default arguments.***<br>
Then we would get an error as.

>><h3>Error:</h3>
SyntaxError: non-default argument follows default argument


```
def sum(a=8,b): #function with non-default followind default value
  print('Sum is :',a+b)

sum(100, 252)
```

> <h3>b.	Python Keyword Argument:</h3>
When we call a function with some values, these values get assigned to the arguments according to their position.

For Example, in earlier function ***sum(a, b)*** when we called it as ***sum(123,523)*** the value ***123*** gets assigned to ***a*** & ***523*** to ***b***. Python allows us to call these functions using keyword arguments, using which we can alter the position or order in which we pass the arguments.

```
def sum(a,b):
  print('A is : {0} \tB is : {1}'.format(a,b))

sum(a=123, b=523)     #arguments passed are in order.
sum(b=56, a=85)       #arguments passed are not in order.
sum(45, b=23)         #1 default, 1 keword argument passed in order
sum(33, 66)           #arguments passed without keys, defaul order.
```

In above example we have 4 different types of function calls, and three different ways with which we can call the user defined function.
1.	In first scenario, we passed both the arguments and these arguments are also in order to the function definition.
2.	In second case, both the arguments are keyword arguments but are not in order.
3.	In third case, we have mix type of arguments here one is positional and another is keyword type argument.
4. For last use case we did not pass any keywards, so they follow the default argument order.

>><h3>Note: </h3>
We can have mix positional arguments with keyword arguments during a function call. But we must keep in mind that keyword arguments must follow positional arguments. Having a positional argument after keyword arguments will result in errors. For example, the function call as follows:

***1. Having a positional argument after keyword arguments will result in errors***

>><h3>Error:</h3>
Error Raised: SyntaxError: positional argument follows keyword argument



><h3>c.	Python Arbitrary Argument:</h3>
Sometimes, we do not know in advance the number of arguments that will be passed into a function. Python allows us to handle this kind of situation through function calls with an arbitrary number of arguments. 

In the function definition, we use an asterisk (*) before the parameter name to denote this kind of argument. Here is an example.

```
  def print_val(*val):          #val as an arbitory argument
    for i in val:               #iterate over values in val
      print('Value is :',i)

  print_val(210,536,85,96,456,53,99,66,33,451)
```

Here, we have called the function with multiple arguments. These arguments get wrapped up into a tuple before being passed into the function. Inside the function, we use a for loop to retrieve all the arguments back.

4. Flush<br>By default, the ***Flush*** is set to ***False***, we can set it to ***True*** as it is a Boolean expression. If flush is set to True, the stream is forcibly flushed.



*   What forcibly flushed means and why we need Flush?<br>
>There are a couple of things to understand here. One is the difference between buffered I/O and unbuffered I/O. The concept is fairly simple - for buffered I/O, there is an internal buffer which is kept. Only when that buffer is full (or some other event happens, such as it reaches a newline) is the output "flushed". With unbuffered I/O, whenever a call is made to output something, it will do this, 1 character at a time.<br>
Most I/O functions fall into the buffered category, mainly for performance reasons: it's a lot faster to write chunks at a time (all I/O functions eventually get down to syscalls of some description, which are expensive.)  Flush lets you manually choose when you want this internal buffer to be written - a call to flush will write any characters in the buffer. Generally, this isn't needed, because the stream will handle this itself. However, there may be situations when you want to make sure something is output before you continue - this is where you'd use a call to flush().<p>

5. Format Print<br>
By using format argument, you can pass as many variables as you want to print. When you pass the variables in the format function you need to specify the index numbers (order in which they are placed inside the format argument) in a predefined string.

image.png

```
a = 'Arya Stark'
b = 18

print("{0} is {1} year's old.".format(a,b))
print("{1} is an Integer, while {0} is a String".format(a,b))
```

Similar to a format argument where your print function acts as a template you can use it to print the values of variables.
* %d – is used as a placeholder for numeric or decimal values.
* %s – is used as a placeholder for string.
* %f - Floating point numbers
* %.\<number of digits>f - Floating point numbers with a fixed number of digits to the right of the dot.
* %x/%X - Integers in hex representation (lowercase/uppercase)
* Dynamic type casting is supported here, so even if you use ***%s*** python will type cast ***int*** to ***str*** while printing output. However, opposite is not True. It will fail to convert a string to an integer and it will result in a Type Error.<p>

```
a = 18
b= 'Arya Stark'

print("%s is %d year's old."%(b,a))
print("%d is an Integer while, %s is a String."%(a,b))
```
